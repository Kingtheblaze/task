## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Load the starter CSV
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. Total evaluate-able rows
total_rows = len(df)

# 2. Find the baseline precision proxy (how many pages are actually declining)
declining_pages = len(df[df['trend_direction'] == 'down'])
decline_rate = (declining_pages / total_rows) * 100

# 3. Check signal density (e.g., how many pages have enough impressions to matter)
high_volume_pages = len(df[df['impressions_90d'] >= 500])

print(f"Total Rows: {total_rows}")
print(f"Pages showing actual decline: {declining_pages} ({decline_rate:.1f}%)")
print(f"Pages with actionable volume (>500 impressions): {high_volume_pages}")
```
